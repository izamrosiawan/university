```python
for col in df_num:
    plt.figure(figsize=(8, 4))
    sns.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.ylabel('Value')
    plt.show()
```

```python
for col in df_num:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    lower_bound = Q1 - 1.5 * (Q3 - Q1)
    upper_bound = Q3 + 1.5 * (Q3 - Q1)

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f'Outliers in {col}:')
    print(f"Q1: {Q1}, Q3: {Q3}")
    print(f"jumlah outlier: {outliers.shape[0]}")
```

```python
df.groupby("Kecamatan")[["Luas genangan", "Tinggi genangan", "Lama genangan"]].mean()
```

```python
hasil_groupby = (
    df.groupby(['Kecamatan', 'Kelurahan', 'Rayon'], as_index=False)
      .agg({
          'Luas genangan': ['mean', 'sum', 'max', 'min'],
          'Tinggi genangan': ['mean', 'sum', 'max', 'min'],
          'Lama genangan': ['mean', 'sum', 'max', 'min'],
          'Skor Luas Genangan': ['mean', 'sum', 'max', 'min'],
          'Skor Tinggi Genangan': ['mean', 'sum', 'max', 'min'],
          'Skor Lama Genangan': ['mean', 'sum', 'max', 'min']
      })
)

hasil_groupby.columns = [
    ''.join(col).strip('') for col in hasil_groupby.columns.values
]

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.colheader_justify', 'center')

print(hasil_groupby.to_string(index=False))
```
