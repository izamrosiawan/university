# Dataset Processing for Image Classification

This notebook processes the fruit/vegetable dataset by splitting the training data into train and validation sets.

**Dataset Structure:**
- Original: All training data in `train/train/` directory
- After processing:
  - `train_test/`: 90% of images for training
  - `valid/`: 10% of images for validation

**Split Ratio:** 90% train, 10% validation

```python
import os
import shutil
import random
import math
```

```python
source_dir = r"C:\Users\LENOVO\OneDrive\Office\GitHub\ss-code-semester-3\infrastruktur-dan-platform-untuk-sains-data\minggu-11\lab\dataset"
#sesuaikan sama lokasi directory
```

```python
train_test_dir_name = "train_test"
valid_dir_name = "valid"
```

```python
path_train_test = os.path.join(source_dir, train_test_dir_name)
path_valid = os.path.join(source_dir, valid_dir_name)
```

```python
def process_dataset(base_path, train_test_path, valid_path):
    # Create destination directories
    os.makedirs(train_test_path, exist_ok=True)
    os.makedirs(valid_path, exist_ok=True)
    
    # Look for the actual training data - it's in train/train/ subdirectory
    train_data_path = os.path.join(base_path, "train", "train")
    
    if not os.path.exists(train_data_path):
        print(f"Error: Training data path not found: {train_data_path}")
        return
    
    # Get class folders from the training data
    class_folders = [item for item in os.listdir(train_data_path) 
                    if os.path.isdir(os.path.join(train_data_path, item))]
    
    print(f"Found {len(class_folders)} classes: {class_folders[:5]}...")  # Show first 5 classes
    
    total_moved_valid = 0
    total_moved_train = 0
    
    for class_name in class_folders:
        src_class_path = os.path.join(train_data_path, class_name)
        
        # Filter hanya mengambil file gambar (Safety Check)
        valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp')
        files = [f for f in os.listdir(src_class_path) if f.lower().endswith(valid_extensions)]
        
        if len(files) == 0:
            print(f"No image files found in {class_name}, skipping...")
            continue
            
        # Set random seed for reproducibility
        random.seed(42)
        random.shuffle(files)
        
        # Split: 10% for validation, 90% for training
        count_valid = int(len(files) * 0.1)
        count_train = len(files) - count_valid
        
        # Create destination class directories
        dest_valid_class = os.path.join(valid_path, class_name)
        dest_train_class = os.path.join(train_test_path, class_name)
        os.makedirs(dest_valid_class, exist_ok=True)
        os.makedirs(dest_train_class, exist_ok=True)
        
        # Move files to validation directory
        for i in range(count_valid):
            filename = files[i]
            src_file = os.path.join(src_class_path, filename)
            dst_file = os.path.join(dest_valid_class, filename)
            shutil.copy2(src_file, dst_file)  # Use copy2 instead of move for safety
        
        # Move files to train_test directory
        for i in range(count_valid, len(files)):
            filename = files[i]
            src_file = os.path.join(src_class_path, filename)
            dst_file = os.path.join(dest_train_class, filename)
            shutil.copy2(src_file, dst_file)  # Use copy2 instead of move for safety
        
        total_moved_valid += count_valid
        total_moved_train += count_train
        
        print(f"Processed {class_name}: {count_train} train, {count_valid} valid")
    
    print(f"\n✅ Processing complete!")
    print(f"Total train images: {total_moved_train}")
    print(f"Total validation images: {total_moved_valid}")
    print(f"Total images processed: {total_moved_train + total_moved_valid}")
```

```python
process_dataset(source_dir, path_train_test, path_valid)
```
