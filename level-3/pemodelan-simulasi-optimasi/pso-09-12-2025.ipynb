```python
import random

Q = 0
N = 100

for i in range(N):
    x = -1 + 2*random.random()
    y = -1 + 2*random.random()
    if x**2 + y**2 <= 1:
        Q += 1

print(Q)
print("Estimated value of π:", 4*Q/N)
```

```python
import numpy as np

Q = 0
N = 100
K = 30
pi_est = []

for j in range(K):
    Q = 0
    for i in range(N):
        x = -1 + 2*random.random()
        y = -1 + 2*random.random()
        if x**2 + y**2 <= 1:
            Q += 1
    pi_est.append(4*Q/N)

mu = np.mean(pi_est)
sig = np.std(pi_est)
print("Mean of π estimates:", mu)
print("Standard deviation of π estimates:", sig)
print(mu - 1.96 * sig)
print(mu + 1.96 * sig)
```
