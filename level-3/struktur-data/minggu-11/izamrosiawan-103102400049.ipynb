```python
# Node untuk BST
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Class BST nya
class BinarySearchTree:
    def __init__(self):
        self.root = None
    
    # Fungsi buat bikin root tree
    def createTree(self, value):
        if self.root == None:
            self.root = Node(value)
        else:
            print("Root sudah ada!")
    
    # Fungsi buat nambahin node
    def addNode(self, value):
        if self.root == None:
            self.root = Node(value)
        else:
            self._tambahNode(self.root, value)
    
    def _tambahNode(self, current, value):
        if value < current.value:
            if current.left == None:
                current.left = Node(value)
            else:
                self._tambahNode(current.left, value)
        elif value > current.value:
            if current.right == None:
                current.right = Node(value)
            else:
                self._tambahNode(current.right, value)
        # kalau sama, ga usah ditambah
    
    # Fungsi buat nampilin tree (inorder)
    def displayTree(self):
        hasil = []
        self._inorder(self.root, hasil)
        return hasil
    
    def _inorder(self, node, hasil):
        if node != None:
            self._inorder(node.left, hasil)
            hasil.append(node.value)
            self._inorder(node.right, hasil)
    
    # Fungsi buat nyari nilai
    def searchNode(self, value):
        return self._cari(self.root, value)
    
    def _cari(self, node, value):
        if node == None:
            return False
        if value == node.value:
            return True
        elif value < node.value:
            return self._cari(node.left, value)
        else:
            return self._cari(node.right, value)
    
    # Fungsi hitung jumlah node
    def countNodes(self):
        return self._hitung(self.root)
    
    def _hitung(self, node):
        if node == None:
            return 0
        return 1 + self._hitung(node.left) + self._hitung(node.right)
    
    # Fungsi cari nilai terkecil
    def getMinValue(self):
        if self.root == None:
            return None
        current = self.root
        while current.left != None:
            current = current.left
        return current.value
    
    # Fungsi cari nilai terbesar
    def getMaxValue(self):
        if self.root == None:
            return None
        current = self.root
        while current.right != None:
            current = current.right
        return current.value
```

```python
bst = BinarySearchTree()

bst.createTree(50)

bst.addNode(30)
bst.addNode(70)
bst.addNode(20)
bst.addNode(40)
bst.addNode(60)
bst.addNode(80)

# Tampilin hasilnya
print("Inorder Traversal:", bst.displayTree())
print("Jumlah Node:", bst.countNodes())
print("Nilai Terkecil:", bst.getMinValue())
print("Nilai Terbesar:", bst.getMaxValue())

# Coba cari beberapa nilai
print("\nCoba cari nilai:")
print("40 ada?", bst.searchNode(40))
print("25 ada?", bst.searchNode(25))
print("80 ada?", bst.searchNode(80))
```
