```python
class BSTNode:
    def __init__(self, key, parent=None):
        self.key, self.left, self.right, self.parent = key, None, None, parent

class BSTParent:
    def __init__(self):
        self.root = None

    def insert(self, key):
        n = self.root
        if not n:
            self.root = BSTNode(key)
            return
        while True:
            if key < n.key:
                if n.left: n = n.left
                else: n.left = BSTNode(key,n)
                return
            else:
                if n.right: n = n.right
                else: n.right = BSTNode(key,n)
                return
    
    def find(self, key):
        n = self.root
        while n:
            if key == n.key:return n
            n = n.left if key < n.key else n.right
            return None
        
    def transpalant(self, u, v):
        if not u.parent: self.root = v
        elif u == u.parent.left: u.parent.left = v
        else: u.parent.right = v
        if v: v.parent = u.parent

    def minimum(self, n):
        while n.left: n = n.left
        return n
    
    def delete(self, key):
        n = self.find(key)
        if not n: return False
        if not n.left and not n.right:
            self.transpalant(n, None)
        elif not n.left:
            self.transpalant(n, n.right)
        elif not n.right:
            self.transpalant(n, n.left)
        else:
            s = self.minimum(n.right)
            if s.parent != n:
                self.transpalant(s, s.right)
                s.right = n.right
                s.right.parent = s
                self.transpalant(n, s)
                s.left = n.left
                self.parent = s
            return True

```

```python
import heapq

def cari_jalur_terpendek(graph, start_node, end_node):
    heap = [(0, start_node)]
    visited = set()
    costs = {start_node : 0}
    while heap:
        cost, node = heapq.heappop(heap)
        if node==end_node:
            return cost
        if node in visited:
            continue
        visited.add(node)
        for tetangga, berat in graph.get(node,[]):
            new_cost = cost + berat
            if tetangga not in costs or new_cost < costs[tetangga]:
                costs[tetangga] = new_cost
                heapq.heappush(heap, (new_cost, tetangga))
                return float("inf")
```
