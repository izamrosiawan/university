```python
# 1. Immutability - data nggak bisa diubah, bikin baru
# tuple adalah data yang immutable (nggak bisa diubah)
x = ("Alice", "Bob", "Charlie")

# fungsi untuk tambah nama ke tuple
def add(d, n):
    # nggak ngubah d yang asli, tapi bikin tuple baru
    return d + (n,)

# panggil fungsi add
y = add(x, "Diana")
# lihat hasilnya: data asli tetap, ada data baru
print("Awal:", x)
print("Baru:", y)
```

```python
# 2. Referential Transparency - fungsi pure, hasil selalu sama
def length(nama):
    # hitung panjang string nama
    return len(nama)

# panggil fungsi dua kali dengan input sama
# hasil pasti sama, nggak akan berubah
print("Panjang 'Alice':", length("Alice"))
print("Panjang 'Alice' lagi:", length("Alice"))  # selalu 5
```

```python
# 3. Higher Order Function - fungsi terima fungsi sebagai parameter
# fungsi map_fn terima 2 parameter: data dan fungsi
def map_fn(d, f):
    # terapkan fungsi f ke setiap item di d
    return tuple(f(i) for i in d)

# fungsi untuk hitung panjang nama
def length(nama):
    # return panjang string nama
    return len(nama)

# panggil map_fn dengan data dari cell 1 (y) dan fungsi length
# hasilnya tuple baru berisi panjang setiap nama
z = map_fn(y, length)
print("Hasil (panjang nama):", z)
```
