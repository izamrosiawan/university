```python
x = (1, 2, 3)

# 1. Immutability
def add(d, n):
    return d + (n,)

y = add(x, 4)
print("Awal:", x)
print("Baru:", y)
```

```python
# 2. Referential Transparency
def sqr(n):
    return n * n

print("Kuadrat 3:", sqr(3))
print("Kuadrat 3 lagi:", sqr(3))
```

```python
# 3. Higher Order Function
def map_fn(d, f):
    return tuple(f(i) for i in d)

z = map_fn(y, sqr)
print("Hasil (kuadrat):", z)
```
