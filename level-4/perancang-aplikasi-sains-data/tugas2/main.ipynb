```python
# Data awal (tuple nggak bisa diubah)
x = (1, 2, 3)

# 1. Immutability - tambah angka tanpa ubah yang asli
def add(d, n):
    # kembalikan tuple baru dengan angka baru ditambah
    return d + (n,)

# panggil fungsi add, simpan hasilnya di y
y = add(x, 4)
print("Awal:", x)  # masih (1, 2, 3)
print("Baru:", y)  # jadi (1, 2, 3, 4)
```

```python
# 2. Referential Transparency - fungsi pure, hasil selalu sama
def sqr(n):
    # kaliin angka sama diri sendiri
    return n * n

# panggil sqr(3) dua kali, hasilnya pasti sama
print("Kuadrat 3:", sqr(3))
print("Kuadrat 3 lagi:", sqr(3))
```

```python
# 3. Higher Order Function - fungsi terima fungsi sebagai parameter
def map_fn(d, f):
    # terapkan fungsi f ke setiap item di d
    return tuple(f(i) for i in d)

# panggil map_fn dengan fungsi sqr, hasilnya tuple baru
z = map_fn(y, sqr)
print("Hasil (kuadrat):", z)
```
