```python
# Tabel Perkalian 1 sampai 10
print("TABEL PERKALIAN 1 sampai 10")
print("-" * 50)

for i in range(1, 11):
    row = ""
    for j in range(1, 11):
        row += f"{i * j:4d}"
    print(f"{i:2d} | {row}")

print("-" * 50)
```

```python
def is_prime(number: int) -> bool:
    """
    Mengecek apakah sebuah angka adalah bilangan prima.
    
    Args:
        number: Angka yang akan dicek
        
    Returns:
        True jika bilangan prima, False jika bukan
    """
    if number < 2:
        return False
    
    if number == 2:
        return True
    
    if number % 2 == 0:
        return False
    
    for i in range(3, int(number**0.5) + 1, 2):
        if number % i == 0:
            return False
    
    return True


# Test fungsi is_prime
print("\nTEST FUNGSI is_prime (1 sampai 30):")
print("-" * 50)

primes = [num for num in range(1, 31) if is_prime(num)]
print(f"Bilangan Prima: {primes}")

print(f"\nTotal bilangan prima dari 1-30: {len(primes)}")
```
